### 3. Write a Python program to check if the given number is Happy Number?
##### ans:

```python
def fun(num):

    res = num; 
    def isHappy(num):    
        r = 0;  
        s = 0
        while(num > 0):    
            r = num%10    
            s += r**2  
            num //= 10    
        return s     

    while(res != 1 and res != 4):    
        res = isHappy(res)    

    if(res == 1):    
        return True
    elif(res == 4):    
        return False
```

### 5. Write a Python program to determine whether the given number is a Harshad Number?
##### ans:

```python
def fun2(num):
    temp = num
    total = 0
    while (temp > 0):
        total = total + (temp%10)
        temp = temp//10
        
    if num%total==0:
        print("Harshad Number")
    else:
        print("Not Harshad Number")
```
